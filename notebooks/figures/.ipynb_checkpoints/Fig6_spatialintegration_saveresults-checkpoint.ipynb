### Figure notebook for integrating MERFISH with MACA data using scANVI and saving results for future plotting

```python
import os
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt
import numpy as np
import tangram as tg
import anndata as ad
import scanpy as sc
import seaborn as sns
from fs import open_fs
from matplotlib_scalebar.scalebar import ScaleBar
from tqdm import tqdm
import scvi
import string

mpl.rcParams.update(mpl.rcParamsDefault) #Reset rcParams to default
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']  # Colors in this style

# Plotting style function (run this before plotting the final figure)
def set_plotting_style():
    plt.style.use('seaborn-paper')
    plt.rc('axes', labelsize=12)
    plt.rc('axes', titlesize=12)
    plt.rc('xtick', labelsize=10)
    plt.rc('ytick', labelsize=10)
    plt.rc('legend', fontsize=10)
    plt.rc('text.latex', preamble=r'\usepackage{sfmath}')
    plt.rc('xtick.major', pad=2)
    plt.rc('ytick.major', pad=2)
    plt.rc('mathtext', fontset='stixsans', sf='sansserif')
    plt.rc('figure', figsize=[10,9])
    plt.rc('svg', fonttype='none')
    
# Function to preprocess raw AnnData object

def preprocess_ad(ad, min_counts):
    ad.layers['counts'] = ad.X.copy()
    sc.pp.filter_cells(ad, min_counts=min_counts)
    sc.pp.normalize_total(ad, target_sum=1e4) # Normalize data
    sc.pp.log1p(ad) # Logarithmize the data
    sc.pp.scale(ad) # Scale each gene to unit variance
```

```python
# Load postprocessed MERFISH and MACA results

# Define the path of the analyzed MERFISH data
dataPathPost = '../../../results'

# Define the experiments
experimentName = ['MsLiver_Cellbound_VZG116_V1_JH_09-18-2021', 
                  'MsKidney_CellBoundary_VZG116_111921']
prefixCountsFilter = ''
suffixCountsFilter = '_FilteredSingleCellCounts.h5ad'

VizgenCountsFilter = []

for i in range(len(experimentName)):
    # Filtered counts per segmented cell
    VizgenCountsFilter.append(ad.read_h5ad(os.path.join(dataPathPost, prefixCountsFilter, 
                                                        experimentName[i] + suffixCountsFilter)))

# Convert all gene names to lower case
for i in range(len(experimentName)):
    VizgenCountsFilter[i].var.index = VizgenCountsFilter[i].var.index.str.lower()
    
# Rename the 5 genes that are inconsistent with MACA (NOTE: MIR205HG doesn't seem to be in MACA at all)
rename_map = {'mir205hg':'4631405k08rik',
             'ackr1':'darc',
             'adgrl4':'eltd1',
             'cavin2':'sdpr',
             'jchain':'igj'}
for i in range(len(experimentName)):
    gene_list = list(VizgenCountsFilter[i].var.index)
    for j in range(len(gene_list)):
        if gene_list[j] in rename_map.keys():
            gene_list[j] = rename_map[gene_list[j]]
    VizgenCountsFilter[i].var.index = gene_list


# Combine into single tissue objects
liver_ind = [0]
kidney_ind = [1]
VizgenLiver_all = []
VizgenKidney_all = []

for i in liver_ind:
    VizgenLiver_all.append(VizgenCountsFilter[i])
    
for i in kidney_ind:
    VizgenKidney_all.append(VizgenCountsFilter[i])

VizgenLiver = VizgenLiver_all[0]
for i in range(len(liver_ind)-1):
    VizgenLiver = VizgenLiver.concatenate(VizgenLiver_all[liver_ind[i+1]])
    
VizgenKidney = VizgenKidney_all[0]
for i in range(len(kidney_ind)-1):
    VizgenKidney = VizgenKidney.concatenate(VizgenKidney_all[kidney_ind[i+1]])
    
# Load raw MACA data (10x)
dataPathMACA = '/mnt/ibm_lg/angela/'
experimentMACA10x = 'tabula-muris-senis-droplet-official-raw-obj.h5ad'
MACA_10x = ad.read(os.path.join(dataPathMACA, experimentMACA10x))

# Load the liver results from Bruce and Abhishek's reannotated map
dataPathReannot = '/mnt/ibm_lg/spatial-seq/MACA/'
nameHep = 'TabulaMuris_3month_Liver_Hepatocyte_Droplet_Raw_reannotated.h5ad'
nameEndo = 'TabulaMuris_1-3month_Liver_Endothelial_Droplet_Raw_reannotated.h5ad'
MACA_10x_hepatocytes = ad.read(os.path.join(dataPathReannot, nameHep))
MACA_10x_endothelial = ad.read(os.path.join(dataPathReannot, nameEndo))

# Keep only the hepatocytes and rename types
MACA_10x_hepatocytes.obs['integrated_manually_annotated'] = MACA_10x_hepatocytes.obs[
    'integrated_manually_annotated'].replace({'Periportal Hepatocytes': 'periportal hepatocyte',
                                              'Perientral Hepatocytes': 'pericentral hepatocyte',
                                              'Inmt High Population': 'other hepatocyte',
                                             })
# Keep only the endothelial cells and rename types
MACA_10x_endothelial.obs['integrated_manually_annotated'] = MACA_10x_endothelial.obs[
    'seurat_clusters_manually_annotated'].replace({'Periportal endothelial': 'periportal endothelial cell',
                                              'Midlobular endothelial': 'other endothelial cell',
                                              'Pericentral endothelial': 'pericentral endothelial cell',
                                              'LSEc (Unknown)': 'other endothelial cell'
                                             })

# Create the combined reannotations into a single dataframe
MACA_10x_liver_reannot_df = pd.concat([MACA_10x_hepatocytes.obs,MACA_10x_endothelial.obs],axis=0)


# Use only the reference datasets that are 1 or 3 months
ind1 = MACA_10x.obs['age'] == '1m' # Indices of 1 month old mice
ind3 = MACA_10x.obs['age'] == '3m' # Indices of 3 month old mice

# Construct liver scRNA-seq reference
# Get the cells corresponding to reannotated hepatocytes and endothelial cells
MACAliver_10x = MACA_10x[np.logical_or(ind1,ind3)] # Load the 1 and 3 month data
MACAliver_10x = MACAliver_10x[MACAliver_10x.obs['tissue'] == 'Liver'].copy()
index_temp = MACAliver_10x.obs.index # Save the original index

# Merge the reannotated cell types
MACAliver_10x.obs = MACAliver_10x.obs.merge(MACA_10x_liver_reannot_df[
    ['cell','integrated_manually_annotated']], how='outer')
MACAliver_10x.obs.index = index_temp # Recover the original index

# Keep only the hepatocytes and endothelial cells
cells_keep = ['periportal hepatocyte','pericentral hepatocyte','other hepatocyte',
              'periportal endothelial cell','pericentral endothelial cell',
                   'other endothelial cell']
MACAliver_10x = MACAliver_10x[MACAliver_10x.obs['integrated_manually_annotated'].isin(cells_keep)]

# Copy over the reannotated labels to the cell_ontology_class column
MACAliver_10x.obs['cell_ontology_class'] = MACAliver_10x.obs['integrated_manually_annotated']

# Add in the non-hepatocyte/endothelial 1 month cells
liver_1month = MACA_10x[ind1]
liver_1month = liver_1month[liver_1month.obs['tissue'] == 'Liver'].copy()
liver_1month = liver_1month[
    np.logical_and(liver_1month.obs['cell_ontology_class'] != 'hepatocyte',
                  liver_1month.obs['cell_ontology_class'] != 'endothelial cell of hepatic sinusoid')]
MACAliver_10x = MACAliver_10x.concatenate(liver_1month)

# Construct kidney scRNA-seq reference
MACAkidney_10x = MACA_10x[ind3]
MACAkidney_10x= MACAkidney_10x[MACAkidney_10x.obs['tissue'] == 'Kidney'].copy()


# Convert genes to lower case
MACAliver_10x.var.index = MACAliver_10x.var.index.str.lower()
MACAkidney_10x.var.index = MACAkidney_10x.var.index.str.lower()

# Select shared gene panel genes only
genes_Vizgen = VizgenCountsFilter[0].var.index
genes_10x = MACAliver_10x.var.index
genes_shared = genes_Vizgen.intersection(genes_10x) # List of shared genes

VizgenLiver = VizgenLiver[:, genes_Vizgen.isin(genes_shared)].copy()
VizgenKidney = VizgenKidney[:, genes_Vizgen.isin(genes_shared)].copy()
MACAliver_10x = MACAliver_10x[:, genes_10x.isin(genes_shared)]
MACAkidney_10x = MACAkidney_10x[:, genes_10x.isin(genes_shared)]

# Final check for uppercase genes in Vizgen data and correct if needed
VizgenLiver.var.index = VizgenLiver.var.index.str.lower()
VizgenKidney.var.index = VizgenKidney.var.index.str.lower()

# Preprocess data
preprocess_ad(VizgenLiver, 80)
preprocess_ad(VizgenKidney, 80)
preprocess_ad(MACAliver_10x, 0)
preprocess_ad(MACAkidney_10x, 0)

print('Data loaded and preprocessed.')
```

```python
# Double check that we are using the following:
# Kidney: 3 month only
# Liver: 3 month hepatocytes, 1-3 month endothelial cells, 1 month other cells
print(MACAkidney_10x.obs['age'].unique())
print(MACAliver_10x.obs['cell_ontology_class'].unique())
print(MACAliver_10x[MACAliver_10x.obs['cell_ontology_class'].isin(['periportal hepatocyte',
                                                                   'pericentral hepatocyte',
                                                                   'other hepatocyte'])].obs['age'].unique())
print(MACAliver_10x[MACAliver_10x.obs['cell_ontology_class'].isin(['periportal endothelial cell',
                                                                   'pericentral endothelial cell',
                                                                   'other endothelial cell'])].obs['age'].unique())
print(MACAliver_10x[MACAliver_10x.obs['cell_ontology_class'].isin(['Kupffer cell',
                                                                   'myeloid leukocyte',
                                                                   'hepatic stellate cell',
                                                                   'B cell',
                                                                   'NK cell',
                                                                   'plasmacytoid dendritic cell'])].obs['age'].unique())
```

```python
# Remap kidney annotations
fileName = '/mnt/ibm_lg/spatial-seq/vizgen-MACA-comparison-paper/results/mousekidney_celltype_relabeling - final.csv'
df_relabel = pd.read_csv(fileName)

# Create dictionary from dataframe
kidneymap = pd.Series(df_relabel['Cell ontology relabel'].values, index=df_relabel['MACA']).to_dict()

# Relabel annotations
MACAkidney_10x.obs['cell_ontology_class'] = MACAkidney_10x.obs['cell_ontology_class'].map(kidneymap)

# Remove "kidney cell" group
MACAkidney_10x = MACAkidney_10x[MACAkidney_10x.obs['cell_ontology_class'] != 'kidney cell']

print('Kidney data re-annotated and cleaned up.')
```

```python
# We'll try integration and label transfer by looking at the subset of MACA that has the same gene panel as MERFISH

# Concatenate the datasets
liver_concat = MACAliver_10x.concatenate(VizgenLiver, batch_key='tech',
                                         batch_categories=['scRNA-seq','MERFISH']).copy()
kidney_concat = MACAkidney_10x.concatenate(VizgenKidney, batch_key='tech',
                                         batch_categories=['scRNA-seq','MERFISH']).copy()

# Use the annotations from the 10x, and treat the MERFISH as unlabeled
liver_concat.obs['celltype_scanvi'] = 'Unknown'
liver_10x_mask = liver_concat.obs['tech'] == 'scRNA-seq'
kidney_concat.obs['celltype_scanvi'] = 'Unknown'
kidney_10x_mask = kidney_concat.obs['tech'] == 'scRNA-seq'

liver_concat.obs['celltype_scanvi'][liver_10x_mask] = liver_concat.obs[
    'cell_ontology_class'][liver_10x_mask].values
kidney_concat.obs['celltype_scanvi'][kidney_10x_mask] = kidney_concat.obs[
    'cell_ontology_class'][kidney_10x_mask].values

# Create the scVI latent space
scvi.model.SCVI.setup_anndata(liver_concat, layer="counts", batch_key="tech")
scvi.model.SCVI.setup_anndata(kidney_concat, layer="counts", batch_key="tech")
vae_liver = scvi.model.SCVI(liver_concat)
vae_kidney = scvi.model.SCVI(kidney_concat)

# Train the liver model
vae_liver.train()

# Train the kidney model
vae_kidney.train()
```

```python
# Register the object and run scANVI

# liver
scvi.model.SCANVI.setup_anndata(
    liver_concat,
    layer="counts",
    batch_key="tech",
    labels_key="celltype_scanvi",
)

# kidney
scvi.model.SCANVI.setup_anndata(
    kidney_concat,
    layer="counts",
    batch_key="tech",
    labels_key="celltype_scanvi",
)

lvae_liver = scvi.model.SCANVI.from_scvi_model(vae_liver, "Unknown", adata=liver_concat)
lvae_kidney = scvi.model.SCANVI.from_scvi_model(vae_kidney, "Unknown", adata=kidney_concat)

lvae_liver.train(max_epochs=20, n_samples_per_label=100)
lvae_kidney.train(max_epochs=20, n_samples_per_label=100)
```

```python
# Get the predicted annotations and latent representation
liver_concat.obs["C_scANVI"] = lvae_liver.predict(liver_concat)
liver_concat.obsm["X_scANVI"] = lvae_liver.get_latent_representation(liver_concat)

kidney_concat.obs["C_scANVI"] = lvae_kidney.predict(kidney_concat)
kidney_concat.obsm["X_scANVI"] = lvae_kidney.get_latent_representation(kidney_concat)

# Calculate the UMAP in the joint latent space
sc.pp.neighbors(liver_concat, use_rep="X_scANVI")
sc.tl.umap(liver_concat)

sc.pp.neighbors(kidney_concat, use_rep="X_scANVI")
sc.tl.umap(kidney_concat)

# Plot the joint UMAPs
sc.pl.umap(liver_concat, color=['tech', 'C_scANVI'])
sc.pl.umap(kidney_concat, color=['tech', 'C_scANVI'])
```

```python
# Plot the spatial plots with transferred annotations
# Get the spatial information
scanvi_liver = liver_concat[liver_concat.obs['tech'] == 'MERFISH'].copy()
scanvi_kidney = kidney_concat[kidney_concat.obs['tech'] == 'MERFISH'].copy()

scanvi_liver.obsm['spatial'] = VizgenLiver.obsm['spatial']
scanvi_kidney.obsm['spatial'] = VizgenKidney.obsm['spatial']

# Plot liver
fig, ax = plt.subplots(figsize=(14,14))
sc.pl.spatial(scanvi_liver, color="C_scANVI", spot_size=100, ax=ax)

# Plot kidney
fig, ax = plt.subplots(figsize=(14,14))
sc.pl.spatial(scanvi_kidney, color="C_scANVI", spot_size=100, ax=ax)
```

```python
# Cell type frequencies
# Now plot the fraction of cell types for each technology
liver_celltypes = liver_concat.obs.loc[:,['tech','C_scANVI']].copy()
liver_celltypes = liver_celltypes.rename(columns={'C_scANVI':'celltype'})

liver_celltype_counts = liver_celltypes.groupby(by=['tech','celltype']).size().unstack().transpose()
liver_celltype_frac = liver_celltype_counts / liver_celltype_counts.sum(axis=0)

liver_celltype_frac.plot.bar(title='liver')
plt.xticks(rotation=45, ha='right')
plt.show()

kidney_celltypes = kidney_concat.obs.loc[:,['tech','C_scANVI']].copy()
kidney_celltypes = kidney_celltypes.rename(columns={'C_scANVI':'celltype'})

kidney_celltype_counts = kidney_celltypes.groupby(by=['tech','celltype']).size().unstack().transpose()
kidney_celltype_frac = kidney_celltype_counts / kidney_celltype_counts.sum(axis=0)

kidney_celltype_frac.plot.bar(title='kidney')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Plot pairwise scatter plots between reference and integration type
fig, ax = plt.subplots(1,2)
liver_celltype_frac.plot.scatter(x='scRNA-seq',y='MERFISH', ax=ax[0], title='liver')
kidney_celltype_frac.plot.scatter(x='scRNA-seq',y='MERFISH', ax=ax[1], title='kidney')
ax[0].plot([0,1],[0,1], 'k--')
ax[1].plot([0,1],[0,1], 'k--')
fig.tight_layout()
plt.show()
```

### Save the results

```python
# Rescale the spatial coordinates from mosaic pixel to micron units
dataPath = '/mnt/ibm_lg/spatial-seq/MERSCOPE/preprocessed_data/'
size_of_one_pixel = {}

for experiment in experimentName:
    transformation_matrix = np.loadtxt(os.path.join(
        dataPath, experiment, 'region_0', 'images', 'micron_to_mosaic_pixel_transform.csv'))

    size_of_one_pixel[experiment] = 1 / transformation_matrix[0,0]

scanvi_liver.obsm['spatial'] = scanvi_liver.obsm['spatial'] * size_of_one_pixel[experimentName[0]]
scanvi_kidney.obsm['spatial'] = scanvi_kidney.obsm['spatial'] * size_of_one_pixel[experimentName[1]]
```

```python
liver_concat.obs = liver_concat.obs.rename(columns={'C_scANVI':'predicted_cell_type'})
kidney_concat.obs = kidney_concat.obs.rename(columns={'C_scANVI':'predicted_cell_type'})
scanvi_liver.obs = scanvi_liver.obs.rename(columns={'C_scANVI':'predicted_cell_type'})
scanvi_kidney.obs = scanvi_kidney.obs.rename(columns={'C_scANVI':'predicted_cell_type'})

saveDir = '../../../results/'

# Save the scVI results
liver_concat.write(saveDir + 'scVI_liver_object.h5ad')
kidney_concat.write(saveDir + 'scVI_kidney_object.h5ad')

# Save the scANVI results
scanvi_liver.write(saveDir + 'scANVI_liver_object.h5ad')
scanvi_kidney.write(saveDir + 'scANVI_kidney_object.h5ad')

print('Results saved in ' + saveDir)
```
