### Figure notebook for integrating MERFISH with MACA data

```python
import os
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt
import numpy as np
import tangram as tg
import anndata as ad
import scanpy as sc
import seaborn as sns
from fs import open_fs
from matplotlib_scalebar.scalebar import ScaleBar
from tqdm import tqdm

mpl.rcParams.update(mpl.rcParamsDefault) #Reset rcParams to default
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']  # Colors in this style

# Plotting style function (run this before plotting the final figure)
def set_plotting_style():
    plt.style.use('seaborn-paper')
    plt.rc('axes', labelsize=12)
    plt.rc('axes', titlesize=12)
    plt.rc('xtick', labelsize=10)
    plt.rc('ytick', labelsize=10)
    plt.rc('legend', fontsize=10)
    plt.rc('text.latex', preamble=r'\usepackage{sfmath}')
    plt.rc('xtick.major', pad=2)
    plt.rc('ytick.major', pad=2)
    plt.rc('mathtext', fontset='stixsans', sf='sansserif')
    plt.rc('figure', figsize=[10,9])
    plt.rc('svg', fonttype='none')
    
# Function to preprocess raw AnnData object

def preprocess_ad(ad, min_counts):
    sc.pp.filter_cells(ad, min_counts=min_counts)
    sc.pp.normalize_total(ad, target_sum=1e4) # Normalize data
    sc.pp.log1p(ad) # Logarithmize the data
    #sc.pp.scale(ad) # Scale each gene to unit variance
```

```python
# Load postprocessed MERFISH and MACA results

# Define the path of the analyzed MERFISH data
dataPathPost = '/mnt/ibm_lg/spatial-seq/MERlin_Analysis/ProcessedResults'

# Define the experiments
experimentName = ['MsLiver_Cellbound_VZG116_V1_JH_09-18-2021', 
                  'MsKidney_CellBoundary_VZG116_111921']
prefixCountsFilter = 'FilteredCounts'
suffixCountsFilter = '_FilteredSingleCellCounts.h5ad'

VizgenCountsFilter = []
QCFilter = []

for i in range(len(experimentName)):
    # Filtered counts per segmented cell
    VizgenCountsFilter.append(ad.read_h5ad(os.path.join(dataPathPost, prefixCountsFilter, 
                                                        experimentName[i] + suffixCountsFilter)))

    # Get filtering criteria
    QCFilter.append(VizgenCountsFilter[i].uns['QC_filter'])
    


# Convert all gene names to lower case
for i in range(len(experimentName)):
    VizgenCountsFilter[i].var.index = VizgenCountsFilter[i].var.index.str.lower()


# Combine into single tissue objects
liver_ind = [0]
kidney_ind = [1]
VizgenLiver_all = []
VizgenKidney_all = []

for i in liver_ind:
    VizgenLiver_all.append(VizgenCountsFilter[i])
    
for i in kidney_ind:
    VizgenKidney_all.append(VizgenCountsFilter[i])

VizgenLiver = VizgenLiver_all[0]
for i in range(len(liver_ind)-1):
    VizgenLiver = VizgenLiver.concatenate(VizgenLiver_all[liver_ind[i+1]])
    
VizgenKidney = VizgenKidney_all[0]
for i in range(len(kidney_ind)-1):
    VizgenKidney = VizgenKidney.concatenate(VizgenKidney_all[kidney_ind[i+1]])
    
# Load raw MACA data (10x)
dataPathMACA = '/mnt/ibm_lg/angela/'
experimentMACA10x = 'tabula-muris-senis-droplet-official-raw-obj.h5ad'
MACA_10x = ad.read(os.path.join(dataPathMACA, experimentMACA10x))

# Select only the tissue-specific cells from the raw datasets
MACAliver_10x= MACA_10x[MACA_10x.obs['tissue'] == 'Liver'].copy()
MACAkidney_10x= MACA_10x[MACA_10x.obs['tissue'] == 'Kidney'].copy()


# Convert genes to lower case
MACAliver_10x.var.index = MACAliver_10x.var.index.str.lower()
MACAkidney_10x.var.index = MACAkidney_10x.var.index.str.lower()

# Select shared gene panel genes only
genes_Vizgen = VizgenCountsFilter[0].var.index
genes_10x = MACAliver_10x.var.index
genes_shared = genes_Vizgen.intersection(genes_10x) # List of shared genes

VizgenLiver = VizgenLiver[:, genes_Vizgen.isin(genes_shared)].copy()
VizgenKidney = VizgenKidney[:, genes_Vizgen.isin(genes_shared)].copy()
MACAliver_10x = MACAliver_10x[:, genes_10x.isin(genes_shared)]
MACAkidney_10x = MACAkidney_10x[:, genes_10x.isin(genes_shared)]

# Preprocess data
preprocess_ad(VizgenLiver, 80)
preprocess_ad(VizgenKidney, 200)
preprocess_ad(MACAliver_10x, 0)
preprocess_ad(MACAkidney_10x, 0)

print('Data loaded and preprocessed.')
```

```python
np.sum(MACAkidney_10x.obs['age'] == '1m')
```

```python
# Define training set of genes from the gene panel for each tissue
ad_Vizgen_liver = VizgenLiver.copy()
ad_Vizgen_kidney = VizgenKidney.copy()
ad_MACA_liver = MACAliver_10x.copy()
ad_MACA_kidney = MACAkidney_10x.copy()

sc.tl.rank_genes_groups(ad_MACA_liver, groupby="cell_ontology_class", use_raw=False)
sc.tl.rank_genes_groups(ad_MACA_kidney, groupby="cell_ontology_class", use_raw=False)

markers_df_liver = pd.DataFrame(ad_MACA_liver.uns["rank_genes_groups"]["names"]).iloc[0:10, :]
markers_df_kidney = pd.DataFrame(ad_MACA_kidney.uns["rank_genes_groups"]["names"]).iloc[0:10, :]

markers_liver = list(np.unique(markers_df_liver.melt().value.values))
markers_kidney = list(np.unique(markers_df_kidney.melt().value.values))
print(str(len(markers_liver)) + ' liver marker genes defined.')
print(str(len(markers_kidney)) + ' kidney marker genes defined.')


# Preprocess with tangram
tg.pp_adatas(ad_MACA_liver, ad_Vizgen_liver, genes=markers_liver)
tg.pp_adatas(ad_MACA_kidney, ad_Vizgen_kidney, genes=markers_kidney)
```

```python
# Find tangram alignment using clusters method
ad_map_liver = tg.map_cells_to_space(ad_MACA_liver, ad_Vizgen_liver,
    #mode='cells',
    mode="clusters",
    cluster_label='cell_ontology_class',  # .obs field w cell types
    #density_prior='rna_count_based',
    density_prior='uniform',
    num_epochs=1000,
    device="cuda:0",
    #device='cpu',
)

print('Liver alignment done.')

ad_map_kidney = tg.map_cells_to_space(ad_MACA_kidney, ad_Vizgen_kidney,
    #mode='cells',
    mode="clusters",
    cluster_label='cell_ontology_class',  # .obs field w cell types
    #density_prior='rna_count_based',
    density_prior='uniform',
    num_epochs=1000,
    device="cuda:0",
    #device='cpu',
)

print('Kidney alignment done.')
```

```python
# Project annotations onto spatial data
tg.project_cell_annotations(ad_map_liver, ad_Vizgen_liver, annotation="cell_ontology_class")
tg.project_cell_annotations(ad_map_kidney, ad_Vizgen_kidney, annotation="cell_ontology_class")


# Plot most likely cell type for each cell in each tissue
pred_liver = ad_Vizgen_liver.obsm['tangram_ct_pred'].copy()
pred_kidney = ad_Vizgen_kidney.obsm['tangram_ct_pred'].copy()

ad_Vizgen_liver.obs['predicted_cell_type'] = pred_liver.idxmax(axis=1)
ad_Vizgen_kidney.obs['predicted_cell_type'] = pred_kidney.idxmax(axis=1)

# Plot liver
fig, ax = plt.subplots(figsize=(14,14))
sc.pl.spatial(ad_Vizgen_liver, color="predicted_cell_type", spot_size=30, ax=ax)

# Plot kidney
fig, ax = plt.subplots(figsize=(14,14))
sc.pl.spatial(ad_Vizgen_kidney, color="predicted_cell_type", spot_size=30, ax=ax)
```

```python
# Plot UMAPs with projected annotations
ad_Vizgen_liver_UMAP = ad_Vizgen_liver.copy()
ad_Vizgen_kidney_UMAP = ad_Vizgen_kidney.copy()

sc.tl.pca(ad_Vizgen_liver_UMAP, svd_solver='arpack') #Compute PCA
sc.pp.neighbors(ad_Vizgen_liver_UMAP, n_neighbors=15, n_pcs=10) #Compute neighborhood graph (use n_neighbors = 12-15?)
sc.tl.umap(ad_Vizgen_liver_UMAP) #Compute UMAP embedding
sc.pl.umap(ad_Vizgen_liver_UMAP, color='predicted_cell_type') 

sc.tl.pca(ad_Vizgen_kidney_UMAP, svd_solver='arpack') #Compute PCA
sc.pp.neighbors(ad_Vizgen_kidney_UMAP, n_neighbors=15, n_pcs=10) #Compute neighborhood graph (use n_neighbors = 12-15?)
sc.tl.umap(ad_Vizgen_kidney_UMAP) #Compute UMAP embedding
sc.pl.umap(ad_Vizgen_kidney_UMAP, color='predicted_cell_type') 
```
