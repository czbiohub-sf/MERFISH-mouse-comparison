```python
import fs
import os
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt
from matplotlib import patches as mpatches
from matplotlib_scalebar.scalebar import ScaleBar
import numpy as np
import seaborn as sns
from skimage.io import imread
import tifffile
from fs import open_fs
import h5py
import string
from merlin.core import dataset

plt.ion()
```

```python
# Load raw MERFISH results
# Define the path of the analyzed data

data_path = '/mnt/ibm_lg/spatial-seq/MERlin_Analysis'

# Define the experiment

experiment_name = 'Msliver_Cellboundary_PM1_12182020_mosaic'

d_set = dataset.MERFISHDataSet(experiment_name)
fs = open_fs(os.path.join(data_path, experiment_name))

print('Dataset loaded.')
```

```python
z_index = 3
fov = 100
transformation_matrix = np.loadtxt(
    os.path.join(data_path, experiment_name, "BlendMosaic", "region_0", 'micron_to_mosaic_pixel_transform.csv'))
size_of_one_pixel = 1 / transformation_matrix[0, 0]
print(transformation_matrix[0, 0], size_of_one_pixel)
```

# Panel B: Fiducial image used for aligning different stacks

```python
fiducial_image = d_set.get_fiducial_image(0, fov)
print(fiducial_image.shape)
print(fiducial_image.dtype)
plt.axis("off")
plt.imshow(fiducial_image, cmap='gray')
```

# Panel A: Input unaligned hybridization image 

```python
hybridization_image = d_set.get_raw_image(
    0, fov, d_set.z_index_to_position(z_index))
print(hybridization_image.shape)
print(hybridization_image.dtype)
plt.axis("off")
plt.imshow(hybridization_image, cmap='gray', vmin=np.min(hybridization_image), vmax=10000)
```

# Panel C: DAPI image

```python
dapi_image = d_set.get_raw_image(
    20, fov, d_set.z_index_to_position(z_index))
print(dapi_image.shape)
print(dapi_image.dtype)
plt.axis("off")
plt.imshow(dapi_image, cmap='gray')
```

# Panel D: POLYT image

```python
polyt_image = d_set.get_raw_image(
    19, fov, d_set.z_index_to_position(z_index))
print(polyt_image.shape)
print(polyt_image.dtype)
plt.axis("off")
plt.imshow(polyt_image, cmap='gray')
```

# Cell bound1 image - not included needs scaling using vmax

```python
cellbound1_image = d_set.get_raw_image(
    18, fov, d_set.z_index_to_position(z_index))
print(cellbound1_image.shape)
print(cellbound1_image.dtype)
plt.axis("off")
plt.imshow(cellbound1_image, cmap='gray')
```

# Panel E: Cell bound2 image

```python
cellbound2_image = d_set.get_raw_image(
    17, fov, d_set.z_index_to_position(z_index))
print(cellbound2_image.shape)
print(cellbound2_image.dtype)
plt.axis("off")
plt.imshow(cellbound2_image, cmap='gray')
```

# Panel F: Cell bound3 image

```python
cellbound3_image = d_set.get_raw_image(
    16, fov, d_set.z_index_to_position(z_index))
print(cellbound3_image.shape)
print(cellbound3_image.dtype)
plt.axis("off")
plt.imshow(cellbound3_image, cmap='gray')
```

```python
mpl.rcParams.update(mpl.rcParamsDefault) #Reset rcParams to default
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']  # Colors in this style

# Plotting style function (run this before plotting the final figure)
def set_plotting_style():
    plt.style.use('seaborn-paper')
    plt.rc('svg', fonttype='none')
```

```python
# Plot the whole figure
heights = [2048] * 2
widths = [2048] * 2

fig_width = 8.  # inches
fig_height = fig_width * sum(heights) / sum(widths)

fig, ax = plt.subplots(2,3, figsize=(fig_width, fig_height),
        gridspec_kw={'height_ratios':heights})
ax[0, 0].axis("off")
ax[0, 0].imshow(hybridization_image, cmap="gray", vmax=10000)
scalebar = ScaleBar(size_of_one_pixel, "um", color='w', box_color='w', box_alpha=0)
ax[0, 0].add_artist(scalebar)

# Panel B
ax[0, 1].axis("off")
ax[0, 1].imshow(fiducial_image, cmap="gray")
scalebar = ScaleBar(size_of_one_pixel, "um", color='w', box_color='w', box_alpha=0)
ax[0, 1].add_artist(scalebar)

                  
# Panel C
ax[0, 2].axis("off")
ax[0, 2].imshow(dapi_image, cmap="gray")
scalebar = ScaleBar(size_of_one_pixel, "um", color='w', box_color='w', box_alpha=0)
ax[0, 2].add_artist(scalebar)

# Panel D
ax[1, 0].axis("off")
ax[1, 0].imshow(polyt_image, cmap="gray")
scalebar = ScaleBar(size_of_one_pixel, "um", color='w', box_color='w', box_alpha=0)
ax[1, 0].add_artist(scalebar)


# Panel E
ax[1, 1].axis("off")
ax[1, 1].imshow(cellbound2_image, cmap="gray")
scalebar = ScaleBar(size_of_one_pixel, "um", color='w', box_color='w', box_alpha=0)
ax[1, 1].add_artist(scalebar)

# Panel F
ax[1, 2].axis("off")
ax[1, 2].imshow(cellbound3_image, cmap="gray")
scalebar = ScaleBar(size_of_one_pixel, "um", color='w', box_color='w', box_alpha=0)
ax[1, 2].add_artist(scalebar)

# Label subpanels
axes = ax.flat
for n, ax in enumerate(axes):
    ax.text(-0.1, 1.1, string.ascii_uppercase[n], transform=ax.transAxes, 
            size=20, weight='bold')
fig.subplots_adjust(wspace=0, hspace=0, left=0, right=1, bottom=0, top=1)
fig.tight_layout()
```

```python
# Export figures
fig.savefig('Fig1_raw_data.pdf')
fig.savefig('Fig1_raw_data.png')
```
